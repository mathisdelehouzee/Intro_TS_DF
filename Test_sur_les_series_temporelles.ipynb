<h1>Seaborn - Matplot en plus stylé</h1>

```python
import seaborn as sns

df_belgium_death = pd.DataFrame(time_series_belgium_reco)
df_belgium_death
```

```python
df_belgium_death = pd.DataFrame(time_series_belgium_death).reset_index() #Ajoute un indexe
df_belgium_death.columns = ['date', 'cases_number'] #Réécris le nom des colonnes
df_belgium_death
```

```python
df_belgium_recovered = pd.DataFrame(time_series_belgium_reco).reset_index()
df_belgium_recovered.columns = ['date','cases_number']
df_belgium_recovered 
```

```python
df_belgium_confirm = pd.DataFrame(time_series_belgium_confirm).reset_index()
df_belgium_confirm.columns = ['date','cases_number']
df_belgium_confirm
```

```python
df_belgium_2 = df_belgium_death.merge(df_belgium_recovered, on='date')
df_belgium_2 = df_belgium_2.merge(df_belgium_confirm, on='date')
df_belgium_2
```

```python
#df_belgium = df_belgium.drop(columns = 'cases_number')
df_belgium_2.columns = ['date', 'case_number_death','case_number_recovered','case_number_confirm']
df_belgium_2
```

```python
df_belgium_2.info()
```

<h3>Maintenant on cherche à caster les données qui sont des objets en int </h3>

```python
#On convertit les colonnes en int
df_belgium_2['case_number_death'] = df_belgium_2['case_number_death'].astype(int) 
df_belgium_2['case_number_recovered'] = df_belgium_2['case_number_recovered'].astype(int)
df_belgium_2['case_number_confirm'] = df_belgium_2['case_number_confirm'].astype(int)
df_belgium_2.info()
```

```python
#On convertit la date en date
df_belgium_2['date'] = pd.to_datetime(df_belgium_2['date'])
type(df_belgium_2)
```

```python
#Permet de faire des courbes
sns.lineplot(x='date', y="case_number_death",data=df_belgium_2)
sns.lineplot(x='date', y="case_number_recovered",data=df_belgium_2)
sns.lineplot(x='date', y="case_number_confirm",data=df_belgium_2)
```
